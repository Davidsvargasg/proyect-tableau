### SABER QUE CLIENTES SE DEVOLVIERON Y EN QUE PASO

```python
df_control['date_time'] = pd.to_datetime(df_control['date_time'])

# Definir el orden de los pasos
step_order = ['start', 'step_1', 'step_2', 'step_3', 'confirm']

# Crear un diccionario para mapear los pasos a números
step_to_num = {step: i for i, step in enumerate(step_order)}

# Añadir una columna numérica para los pasos
df_control['step_num'] = df_control['web_steps'].map(step_to_num)

# Ordenar el DataFrame
df_control = df_control.sort_values(by=['client_id', 'visitor_id', 'visit_id', 'date_time'])

# Crear una función para detectar devoluciones
def detect_devolutions(group):
    devolution_points = []
    previous_step = -1
    for index, row in group.iterrows():
        if row['step_num'] < previous_step:
            devolution_points.append((row['client_id'], row['web_steps'], row['date_time']))
        previous_step = row['step_num']
    return devolution_points

# Aplicar la función al DataFrame agrupado por client_id, visitor_id, visit_id
devolutions_control = df_control.groupby(['client_id', 'visitor_id', 'visit_id']).apply(detect_devolutions).reset_index(drop=True)

# Filtrar las devoluciones
devolutions_control = [item for sublist in devolutions_control for item in sublist]

# Convertir las devoluciones en un DataFrame
devolutions_control= pd.DataFrame(devolutions_control, columns=['client_id', 'web_steps', 'date_time'])

# Mostrar el DataFrame resultante
devolutions_control
```

## HIPOTESIS

```python
import scipy.stats as st
```

#### HIPOTESIS 1

```python
df_female = df1[df1["gender"]=="F"]["balance"]
df_male = df1[df1["gender"]=="M"]["balance"]
```

```python
#Set the hypothesis

#H0: mu_balance male = mu_balance female
#H1: mu_balance male != mu_balance female

#Significance level -> 0.05
alpha = 0.05

stat, p_value = st.ttest_ind(df_male,df_female, equal_var=False)
print ("stat: "), stat
print ("p_value: ", p_value)

if p_value > alpha:
    print("We are not able to reject the null hypothesis")
else:
    print("We reject the null hypotesis")
```
